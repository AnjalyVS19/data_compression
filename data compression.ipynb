```python
m_noisy=np.asmatrix(np.random.uniform(low=0,high=0.7,size=(250,150)))
m_noisy
```

```python
plt.imshow(m_noisy)
```

```python
m_noisy=m+m_noisy
plt.imshow(m_noisy)
```

```python
from scipy.linalg import svd
u,s,v=svd(m_noisy)
```

```python
print(u.shape,s.shape,v.shape)
```

```python
plt.subplot(1,5,1)
plt.title('input image')
plt.imshow(m_noisy)
for i in range(1,5):
    print(i)
    sig=mat(np.eye(i)*s[:i])
    print(sig)
    u_reduced=u[:,:i]
    v_reduced=v[:i,:]
    m_red=u_reduced*sig*v_reduced
    plt.subplot(1,5,i+1)
    plt.imshow(m_red)
    plt.title(i)
```
